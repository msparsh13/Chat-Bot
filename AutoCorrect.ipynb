```python
def deleteword(word ):
    split= [(word[:i] , word[i:]) for i in range(len(word)) ]
    del_list =[( a[0]+a[1][1:] ) for a in split ]
    return del_list
```

```python
alphabet='abcdefghijklmopqrstuvwxyz'
def replaceword(word , verbose=False):
    split= [(word[:i] , word[i:]) for i in range(len(word)) ]
    replace_list = [(b[0]+a+b[1][1:]) for a in alphabet for b in split ]
    return replace_list
```

```python
replaceword('hating')
```

```python
def switchword(word , verbose=False ):
    split = [(word[:i] , word[i:]) for i in range(1 , len(word))]
    switch_list = [a[0][:len(a[0])-1]+a[1][0]+a[0][len(a[0])-1:]+a[1][1:] for a in split]
    return switch_list
```

```python
def insertword(word , verbose=False):
    split= [(word[:i] , word[i:]) for i in range(len(word)) ]
    insert_list = [b[0]+a+b[1] for a in alphabet for b in split ]
    return insert_list
```

```python

num_enc_tokens = 38
num_dec_tokens = 39 
latent_dim=50


encoder_inputs = Input(shape=(None,num_enc_tokens))
encoder =GRU(latent_dim , return_state=True , dropout=0.2 , return_sequences=True)
encoder_outputs , state = encoder(encoder_inputs)
encoder_states=state

decoder_inputs = Input(shape=(None,num_dec_tokens))
decoder_lstm = GRU(latent_dim,return_sequences=True , return_state=True)
decoder_ouputs  , _= decoder_lstm(decoder_inputs,initial_state = encoder_states )



decoder_dense = Dense(num_dec_tokens, activation='softmax')
decoder_ouputs  = decoder_dense(decoder_ouputs)

model = Model([encoder_inputs,decoder_inputs],decoder_ouputs)
model.compile(optimizer='adam', loss='categorical_crossentropy',
                  metrics=['accuracy'] )
model.summary()
```

```python
h=model.fit([encoder_input_data,decoder_input_data],decoder_target_data,
          validation_split = 0.2 ,
            epochs=100
         )
```

```python
encoder_model = Model(encoder_inputs,encoder_states)

decoder_state_input_h = Input(shape=(latent_dim,))
decoder_states_inputs = [decoder_state_input_h]
decoder_outputs,state_h= decoder_lstm(
        decoder_inputs,initial_state = decoder_states_inputs
)
decoder_states = [state_h]
decoder_outputs = decoder_dense(decoder_outputs)
decoder_model = Model(
    [decoder_inputs] + decoder_states_inputs,
    [decoder_outputs] + decoder_states
)
```

```python
def decode_sequence(input_seq):
    state_value = encoder_model.predict(input_seq)

    target_word = np.zeros((1, 1, num_dec_tokens))

    target_word[0, 0, char2idx[' ']] = 1.


    stop_condition = False
    decoded_sentence = ''
    while not stop_condition:
        output_tokens, h = decoder_model.predict(
            [target_word] +[ state_value])
        sampled_token_index = np.argmax(output_tokens[0, -1, :])
        sampled_char = idx2char[sampled_token_index]
        decoded_sentence += sampled_char
        if (sampled_char == '/n' or
           len(decoded_sentence) > max_dec_len):
            stop_condition = True
        target_word = np.zeros((1, 1, num_dec_tokens))
        target_word[0, 0, sampled_token_index] = 1.
        state_value = [h]
    return decoded_sentence

for seq_index in range(440 , 500):
    input_seq = encoder_input_data[seq_index: seq_index + 1]
    decoded_sentence = decode_sequence(input_seq)
    print('-')
    print('Wrong sentence:', df[0][seq_index])
    print('Corrected sentence:', decoded_sentence)
    print('Ground Truth:',df[1][seq_index])
```
